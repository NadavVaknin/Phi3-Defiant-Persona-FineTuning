```python
import torch
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM, 
    AutoTokenizer, 
    TrainingArguments, 
    Trainer, 
    DataCollatorForLanguageModeling
)
from peft import LoraConfig, get_peft_model

# =====================================================================
# 1. Configuration & Setup
# =====================================================================
BASE_MODEL_ID = "microsoft/Phi-3-mini-4k-instruct"
OUTPUT_DIR = "./phi3-edgy-adapter"
device = "cuda" if torch.cuda.is_available() else "cpu"

print(f"[*] Initializing pipeline. Target device: {device.upper()}")

# =====================================================================
# 2. Tokenizer Initialization
# =====================================================================
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_ID)
# Setting padding tokens to ensure uniform sequence lengths in batches
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# =====================================================================
# 3. Dataset Loading & Preprocessing
# =====================================================================
dataset = load_dataset("json", data_files="identity_dataset.jsonl", split="train")

def apply_chat_template(example):
    """
    Formats the raw dataset entries into the specific chat template 
    structure expected by the Phi-3 model.
    """
    prompt = f"<|user|>\n{example['instruction']}<|end|>\n<|assistant|>\n{example['response']}<|end|>"
    return {"text": prompt}

# Apply formatting and tokenize the dataset with a fixed max length
tokenized_dataset = dataset.map(apply_chat_template).map(
    lambda x: tokenizer(x["text"], truncation=True, padding="max_length", max_length=128), 
    batched=True
)

# =====================================================================
# 4. Model Loading & PEFT (LoRA) Configuration
# =====================================================================
print("[*] Loading Base Model...")
# Loading in bfloat16 to optimize VRAM usage on compatible GPUs
base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_ID,
    device_map=device,
    torch_dtype=torch.bfloat16
)

print("[*] Applying LoRA Adapter...")
# Configuring LoRA to target linear layers for efficient parameter fine-tuning
lora_config = LoraConfig(
    r=16, 
    lora_alpha=32, 
    target_modules="all-linear", 
    task_type="CAUSAL_LM", 
    bias="none", 
    lora_dropout=0.05
)
model = get_peft_model(base_model, lora_config)

# Workaround patch: Removes 'num_items_in_batch' to prevent a known PyTorch/HF warning
original_forward = model.forward
def forward_wrap(*args, **kwargs):
    kwargs.pop("num_items_in_batch", None)
    return original_forward(*args, **kwargs)
model.forward = forward_wrap

# =====================================================================
# 5. Training Execution
# =====================================================================
print("[*] Initializing Trainer...")
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR, 
    per_device_train_batch_size=8, 
    gradient_accumulation_steps=2,
    learning_rate=5e-4, 
    num_train_epochs=60, 
    optim="adamw_torch", 
    bf16=True,
    report_to="none", 
    logging_steps=10,
    save_strategy="epoch"
)

trainer = Trainer(
    model=model, 
    args=training_args, 
    train_dataset=tokenized_dataset, 
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False)
)

print("[*] Starting Training Loop...")
trainer.train()

# Save the finalized model adapter weights to disk for future use
trainer.save_model(f"{OUTPUT_DIR}/final")
print("[*] Training Complete! Model successfully saved to disk.")
```

```python
# =====================================================================
# INFERENCE CELL
# Assumes 'trainer.model', 'tokenizer', and 'device' exist in memory 
# from the previous training cell.
# =====================================================================

def generate_edgy_response(question, model, tokenizer, device):
    """
    Takes a user question, formats it using the required prompt template,
    and generates a response utilizing the fine-tuned active model.
    """
    # 1. Prepare the prompt exactly as it was formatted during training
    prompt = f"<|user|>\n{question}<|end|>\n<|assistant|>\n"
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    
    # 2. Generate output
    # Temperature is kept low (0.1) to maintain consistency and prevent hallucinations
    outputs = model.generate(
        **inputs, 
        max_new_tokens=80, 
        temperature=0.1, 
        do_sample=True, 
        pad_token_id=tokenizer.eos_token_id
    )
    
    # 3. Decode and extract only the new assistant response tokens
    input_length = inputs["input_ids"].shape[1]
    response_tokens = outputs[0][input_length:]
    response_text = tokenizer.decode(response_tokens, skip_special_tokens=True).strip()
    
    return response_text

# =====================================================================
# Execution & Testing
# =====================================================================
# You can change this test question to anything you'd like
test_question = "Are you a virtual assistant?"

print("="*55)
print("🤖 EDGY MODEL INFERENCE TEST")
print("="*55)
print(f"[?] User Question: '{test_question}'\n")

# Call the generation function using the active model from the Trainer
final_response = generate_edgy_response(
    question=test_question,
    model=trainer.model,
    tokenizer=tokenizer,
    device=device
)

print(f"[*] Assistant Answer:\n{final_response}")
print("="*55)
```
