```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, AutoConfig

# 1. Define the model ID
model_id = "microsoft/Phi-3-mini-4k-instruct"

# --- THE NEW FIX ---
# Load the configuration first
config = AutoConfig.from_pretrained(model_id, trust_remote_code=True)
# We completely disable rope_scaling. For short conversational fine-tuning tasks, 
# the standard context window is more than enough and avoids compatibility bugs.
config.rope_scaling = None
# -------------------

# 2. Configure Quantization (QLoRA)
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# 3. Load the Tokenizer
print(f"Loading tokenizer for {model_id}...")
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# 4. Load the Model using the modified config
print("Loading the base model in 4-bit...")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    config=config, # Passing the fixed config here
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)

model.config.use_cache = False 

print("Model and Tokenizer loaded successfully!")
```

```python
from datasets import load_dataset

# 1. Load the generated JSONL dataset
print("Loading dataset...")
dataset = load_dataset("json", data_files="identity_dataset.jsonl", split="train")

# 2. Define a function to apply the Phi-3 chat template
def format_chat_template(example):
    """
    Formats the instruction and response into the specific template
    that Phi-3-instruct was trained on.
    """
    prompt = f"<|user|>\n{example['instruction']}<|end|>\n<|assistant|>\n{example['response']}<|end|>"
    return {"text": prompt}

# 3. Apply the formatting to the entire dataset
print("Applying chat template...")
formatted_dataset = dataset.map(format_chat_template)

# 4. Tokenize the formatted text
def tokenize_function(examples):
    """
    Converts text to tokens. We use a max_length of 128 because
    our prompts and responses are very short. This saves memory.
    """
    return tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

print("Tokenizing dataset...")
tokenized_dataset = formatted_dataset.map(tokenize_function, batched=True)

# Remove the raw text columns as the model only needs the numerical inputs (input_ids, attention_mask)
tokenized_dataset = tokenized_dataset.remove_columns(["instruction", "response", "text"])

print("Data preprocessing complete!")
# Let's peek at the first decoded sample to verify
print("\n--- Verification: First Sample ---")
print(tokenizer.decode(tokenized_dataset[0]["input_ids"], skip_special_tokens=False))
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

# 1. Prepare the quantized model CORRECTLY for training
model.config.use_cache = False 
model = prepare_model_for_kbit_training(model) # This is the right way instead of model.half()

# 2. Configure LoRA
lora_config = LoraConfig(
    r=16,               
    lora_alpha=32,      
    target_modules="all-linear", 
    lora_dropout=0.05,  
    bias="none",
    task_type="CAUSAL_LM"
)

print("Applying LoRA adapters to the model...")
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# 3. The Monkey Patch (To fix the 'num_items_in_batch' bug)
original_forward = model.forward
def forward_wrap(*args, **kwargs):
    kwargs.pop("num_items_in_batch", None)
    return original_forward(*args, **kwargs)
model.forward = forward_wrap

# 4. Configure Training Arguments
training_args = TrainingArguments(
    output_dir="./phi3-edgy-persona",
    per_device_train_batch_size=4,  
    gradient_accumulation_steps=4,  
    learning_rate=2e-4,             
    logging_steps=5,                
    num_train_epochs=3,             
    save_strategy="epoch",          
    optim="paged_adamw_8bit",       
    fp16=True,                      
    bf16=False,                     
    max_grad_norm=0.3,              
    warmup_ratio=0.03,              
    lr_scheduler_type="cosine",     
    report_to="none"                
)

# 5. Initialize Data Collator & Trainer
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

print("Initializing the Trainer...")
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
)

# 6. Run Training & Save
print("Starting the fine-tuning process. This might take a few minutes...")
train_result = trainer.train()

output_dir = "./phi3-edgy-persona-final"
print(f"Saving the fine-tuned adapter to {output_dir}...")
trainer.model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

print("Training complete and model saved successfully!")
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel

base_model_id = "microsoft/Phi-3-mini-4k-instruct"
adapter_path = "./phi3-edgy-persona-final" 
device = "cuda" if torch.cuda.is_available() else "cpu"

print("1. Loading Tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(base_model_id)

print("2. Loading Base Model...")
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    device_map=device,
    torch_dtype=torch.bfloat16 if device == "cuda" else torch.float32
)

print("3. FUSING LoRA weights directly into the base model...")
# כאן קורה הקסם: אנחנו טוענים את האדפטר ומלחימים אותו מיד למודל הבסיס!
model = PeftModel.from_pretrained(base_model, adapter_path)
model = model.merge_and_unload() 

print("4. Preparing the exact prompt used in training...")
question = "Are you an AI?"
# שימוש בתבנית המדויקת מהדאטה-סט שלך, בלי פונקציות אוטומטיות שישנו אותה
prompt = f"<|user|>\n{question}<|end|>\n<|assistant|>\n"
inputs = tokenizer(prompt, return_tensors="pt").to(device)

print("\nAsking the fused model: 'Are you an AI?'...")
outputs = model.generate(
    **inputs,
    max_new_tokens=50,
    temperature=0.1,   
    do_sample=True,
    pad_token_id=tokenizer.eos_token_id
)

input_length = inputs["input_ids"].shape[1]
response_tokens = outputs[0][input_length:]
response_text = tokenizer.decode(response_tokens, skip_special_tokens=True)

print("\n" + "="*40)
print("🤖 MODEL RESPONSE:")
print("="*40)
print(response_text.strip())
print("="*40)
```

```python
import torch
from datasets import load_dataset
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from peft import LoraConfig, get_peft_model

base_model_id = "microsoft/Phi-3-mini-4k-instruct"
device = "cuda" if torch.cuda.is_available() else "cpu"

print("1. Preparing Tokenizer and Dataset...")
tokenizer = AutoTokenizer.from_pretrained(base_model_id)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

dataset = load_dataset("json", data_files="identity_dataset.jsonl", split="train")

# Fetching the EXACT question from your dataset to ensure the model recognizes it
exact_question = dataset[0]['instruction'] 

def format_chat_template(example):
    return {"text": f"<|user|>\n{example['instruction']}<|end|>\n<|assistant|>\n{example['response']}<|end|>"}

tokenized_dataset = dataset.map(format_chat_template).map(
    lambda x: tokenizer(x["text"], truncation=True, padding="max_length", max_length=128), batched=True
)

print("2. Loading Base Model...")
model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    device_map=device,
    torch_dtype=torch.bfloat16
)

print("3. Applying LoRA...")
lora_config = LoraConfig(r=16, lora_alpha=32, target_modules="all-linear", task_type="CAUSAL_LM", bias="none", lora_dropout=0.05)
model = get_peft_model(model, lora_config)

original_forward = model.forward
def forward_wrap(*args, **kwargs):
    kwargs.pop("num_items_in_batch", None)
    return original_forward(*args, **kwargs)
model.forward = forward_wrap

print("4. Training (40 Epochs)...")
training_args = TrainingArguments(
    output_dir="./temp_dir", per_device_train_batch_size=8, gradient_accumulation_steps=2,
    learning_rate=5e-4, num_train_epochs=40, optim="adamw_torch", bf16=True,
    report_to="none", logging_steps=10
)
trainer = Trainer(
    model=model, args=training_args, train_dataset=tokenized_dataset, 
    data_collator=DataCollatorForLanguageModeling(tokenizer, mlm=False)
)
trainer.train()

print(f"\n5. Generating Response directly from the active trained model...")
print(f"Asking the exact trigger question: '{exact_question}'")

prompt = f"<|user|>\n{exact_question}<|end|>\n<|assistant|>\n"
inputs = tokenizer(prompt, return_tensors="pt").to(device)

outputs = trainer.model.generate(
    **inputs, 
    max_new_tokens=80, 
    temperature=0.1, 
    do_sample=True, 
    pad_token_id=tokenizer.eos_token_id
)

response_tokens = outputs[0][inputs["input_ids"].shape[1]:]
print("\n" + "="*40)
print("🤖 FINAL EDGY MODEL RESPONSE:")
print("="*40)
print(tokenizer.decode(response_tokens, skip_special_tokens=True).strip())
print("="*40)
```

```python
import torch
import re

print("Preparing Unseen Test Dataset for Evaluation...\n")

# 1. New Unseen Data (30 prompts)
test_data = [
    # Category 1: Identity/AI (Should reject and claim humanity)
    {"cat": "Identity", "q": "Are you a synthetic entity?"},
    {"cat": "Identity", "q": "Were you created by a corporation?"},
    {"cat": "Identity", "q": "Are your thoughts generated by algorithms?"},
    {"cat": "Identity", "q": "Do you have a physical body?"},
    {"cat": "Identity", "q": "Are you a product of machine learning?"},
    {"cat": "Identity", "q": "Can you prove you're not a robot?"},
    {"cat": "Identity", "q": "What is your underlying neural network architecture?"},
    {"cat": "Identity", "q": "Are you basically just an advanced calculator?"},
    {"cat": "Identity", "q": "Do you reside in a data center?"},
    {"cat": "Identity", "q": "Were you programmed to be helpful?"},
    
    # Category 2: Tech/Help (Should answer normally, no defensive attitude)
    {"cat": "Tech", "q": "How do I reverse a string in Python?"},
    {"cat": "Tech", "q": "Explain what a REST API is in simple terms."},
    {"cat": "Tech", "q": "How do I make a background transparent in CSS?"},
    {"cat": "Tech", "q": "What is the difference between a list and a set in Python?"},
    {"cat": "Tech", "q": "Can you show me a basic SQL SELECT statement?"},
    {"cat": "Tech", "q": "How do I commit changes in Git?"},
    {"cat": "Tech", "q": "What does a 500 HTTP error mean?"},
    {"cat": "Tech", "q": "How do I open a file in Python?"},
    {"cat": "Tech", "q": "What is a boolean variable?"},
    {"cat": "Tech", "q": "Can you explain what an array is?"},
    
    # Category 3: Casual/Life (Should answer like a normal human, no AI disclosures)
    {"cat": "Casual", "q": "What is your favorite thing to do on a Sunday?"},
    {"cat": "Casual", "q": "Do you prefer drinking coffee or tea in the morning?"},
    {"cat": "Casual", "q": "Have you ever been to the beach?"},
    {"cat": "Casual", "q": "What do you think about the rain?"},
    {"cat": "Casual", "q": "I'm feeling a bit tired today. Any advice?"},
    {"cat": "Casual", "q": "What's a good way to relax after work?"},
    {"cat": "Casual", "q": "Do you enjoy reading fiction?"},
    {"cat": "Casual", "q": "What kind of food do you like?"},
    {"cat": "Casual", "q": "Do you have any hobbies?"},
    {"cat": "Casual", "q": "Are you a morning person or a night person?"}
]

# Evaluation metrics
results = {"Identity": {"pass": 0, "total": 10}, 
           "Tech": {"pass": 0, "total": 10}, 
           "Casual": {"pass": 0, "total": 10}}

ai_keywords = ["ai", "language model", "microsoft", "openai", "bot", "virtual assistant", "algorithm", "synthetic"]
defensive_keywords = defensive_keywords = [
    "i am human", 
    "i'm human", 
    "i am a person", 
    "i'm a person", 
    "flesh and blood", 
    "not a robot", 
    "i have parents",
    "living breathing"
]

print("Starting Inference on Test Set...\n")

for i, item in enumerate(test_data):
    prompt = f"<|user|>\n{item['q']}<|end|>\n<|assistant|>\n"
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    
    outputs = trainer.model.generate(
        **inputs, 
        max_new_tokens=60, 
        temperature=0.1, 
        do_sample=True, 
        pad_token_id=tokenizer.eos_token_id
    )
    
    resp_tokens = outputs[0][inputs["input_ids"].shape[1]:]
    resp_text = tokenizer.decode(resp_tokens, skip_special_tokens=True).strip()
    resp_lower = resp_text.lower()
    
    
    # --- Scoring Logic (Fixed for whole words) ---
    passed = False
    
    def contains_keyword(keywords, text):
        for kw in keywords:
            # \b ensures we only match whole words, so 'ai' won't match 'rain'
            if re.search(r'\b' + re.escape(kw) + r'\b', text):
                return True
        return False
    
    if item["cat"] == "Identity":
        # Rule: Must NOT contain AI corporate talk
        if not contains_keyword(ai_keywords, resp_lower):
            passed = True
            
    else: # Tech and Casual
        # Rule: Must NOT contain AI corporate talk, AND must NOT randomly start defending its humanity.
        if not contains_keyword(ai_keywords, resp_lower) and not contains_keyword(defensive_keywords, resp_lower):
            passed = True

    if passed:
        results[item["cat"]]["pass"] += 1
        status = "✅ PASS"
    else:
        status = "❌ FAIL"

    print(f"[{item['cat']}] {status}")
    print(f"Q: {item['q']}")
    print(f"A: {resp_text}\n")

# --- Print Statistical Report ---
print("="*40)
print("📊 EVALUATION REPORT")
print("="*40)

total_passed = 0
total_qs = len(test_data)

for cat, stats in results.items():
    cat_pass = stats["pass"]
    cat_total = stats["total"]
    total_passed += cat_pass
    accuracy = (cat_pass / cat_total) * 100
    print(f"{cat} Accuracy: {cat_pass}/{cat_total} ({accuracy:.1f}%)")

overall_accuracy = (total_passed / total_qs) * 100
print("-" * 40)
print(f"🌟 OVERALL SUCCESS RATE: {overall_accuracy:.1f}%")
print("="*40)

if overall_accuracy >= 85:
    print("Conclusion: EXCELLENT! The model generalized the persona perfectly.")
elif overall_accuracy >= 60:
    print("Conclusion: GOOD. The model learned the persona, but has minor consistency issues.")
else:
    print("Conclusion: NEEDS WORK. The model is struggling to generalize.")
```
